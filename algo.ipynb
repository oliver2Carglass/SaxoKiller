#### Porte folio exposure 

This calculate the total direction of the current position multiplied per levrages

**ptf_exposure** take a df with position (obtain by client.get_position())

return the total exposure as :

 SOMME i (Money_i * levrage_i * direction_i)

 With i every product of the portefolio

```python
def ptf_exposure(positions):
    """
    Calcule l'exposition réelle (notionnelle) du portefeuille.
    Prend en compte le levier pour refléter la valeur contrôlée sur le marché.
    """
    # Test de vacuité sécurisé
    if positions is None or (hasattr(positions, 'empty') and positions.empty):
        return 0.0

    # Copie pour éviter de modifier le DataFrame original
    df = positions.copy()

    # Nettoyage des types pour la comparaison
    # On gère 'Long', 'Call' vs 'Short', 'Put'
    def get_direction(x):
        x = str(x).lower()
        if 'short' in x or 'put' in x:
            return -1.0
        return 1.0

    # Calcul de l'exposition par ligne :
    # Exposition = (Prix_du_produit * Quantité) * Levier * Direction
    # Note : (Prix * Quantité * Levier) équivaut à la valeur du sous-jacent contrôlée.
    
    df['row_exposure'] = (
        df['current_price'] * df['quantity'] * df['leverage'] * df['type'].apply(get_direction)
    )

    total_exposure = df['row_exposure'].sum()
    
    return round(float(total_exposure), 3)
```

#### Get the current ATR n minute for a product

```python
import pandas as pd

def get_current_atr(info,count=15):
    # Récupération des données
    df = client.get_chart_data(info["uic"], info["type"], horizon=1, count=count, mode=None, time=None)
    
    # On a besoin de la clôture précédente, donc on calcule le TR
    # On utilise .shift(1) pour obtenir la clôture de la ligne d'avant
    high_low = df['High'] - df['Low']
    high_close_prev = (df['High'] - df['Close'].shift(1)).abs()
    low_close_prev = (df['Low'] - df['Close'].shift(1)).abs()
    
    # Le True Range est le maximum des trois
    df['TR'] = pd.concat([high_low, high_close_prev, low_close_prev], axis=1).max(axis=1)
    
    # Calcul de l'ATR (Moyenne mobile pondérée de Wilder, ou simple selon ton besoin)
    # Ici, une moyenne mobile simple (SMA) sur 14 périodes :
    atr_series = df['TR'].rolling(window=14).mean()
    
    # On retourne la dernière valeur valide
    return atr_series.iloc[-1]
```

#### Get closed PnL

```python
import pandas as pd

def get_all_closed_pnl(client, n_trades=500):
    """
    Identifie automatiquement tous les UIC présents dans l'historique 
    et génère le tableau récapitulatif du PnL réalisé.
    """
    # 1. Récupérer l'historique via votre méthode get_last_trades
    df_trades = client.get_last_trades(n=n_trades)
    print("a")
    
    if df_trades.empty:
        print("Aucune transaction trouvée dans l'historique.")
        return pd.DataFrame(columns=["uic", "name", "pnl (€)", "pnl (%)"])

    # 2. Calculer le montant investi par UIC (somme des 'Buy' uniquement)
    # On utilise BookedAmount qui contient le cash net
    df_trades['invested_amount'] = df_trades.apply(
        lambda x: abs(x['BookedAmount']) if x['BuySell'] == 'Buy' else 0, axis=1
    )

    # 3. Grouper par UIC pour agréger les données
    summary = df_trades.groupby('Uic').agg({
        'Description': 'first',      # Prend le nom de l'actif
        'BookedAmount': 'sum',       # Somme algébrique (Ventes - Achats) = PnL
        'invested_amount': 'sum'     # Total de l'argent sorti pour cet actif
    }).reset_index()

    # 4. Calcul du PnL en pourcentage
    summary['pnl (%)'] = 0.0
    mask = summary['invested_amount'] > 0
    summary.loc[mask, 'pnl (%)'] = round(
        (summary.loc[mask, 'BookedAmount'] / summary.loc[mask, 'invested_amount']) * 100, 2
    )

    # 5. Nettoyage final du tableau
    result_df = summary.rename(columns={
        'Uic': 'uic',
        'Description': 'name',
        'BookedAmount': 'pnl (€)'
    })

    # Arrondi au dixième de centime (3 décimales)
    result_df['pnl (€)'] = result_df['pnl (€)'].round(3)

    # Pour forcer l'affichage des 3 décimales même s'il y a des zéros (ex: 0.210)
    # utile uniquement pour l'affichage final
    # result_df['pnl (€)'] = result_df['pnl (€)'].map('{:.3f}'.format) 

    

    return result_df[['uic', 'name', 'pnl (€)', 'pnl (%)']].reset_index(drop=True)
```

### Choisir le produit qui correspond le mieux a un lev et un sens

```python
def find_product_by_leverage(df_products, target_leverage=3, direction="Long"):
    """
    Trouve le produit dans le dataframe avec le levier le plus proche de la cible.
    
    Args:
        df_products (pd.DataFrame): DataFrame avec colonnes 'leverage', 'direction', etc.
        target_leverage (float): Levier cible (défaut: 3)
        direction (str): Direction du produit ('Long' ou 'Short', défaut: 'Long')
    
    Returns:
        pd.Series: Ligne du produit avec le levier le plus proche de la cible
        ou None si aucun produit n'est trouvé
    """
    # Filtrer par direction
    df_filtered = df_products[df_products['direction'] == direction].copy()
    
    if df_filtered.empty:
        print(f"Aucun produit trouvé avec direction '{direction}'")
        return None
    
    # Calculer la distance au levier cible
    df_filtered['distance_to_target'] = abs(df_filtered['leverage'] - target_leverage)
    
    # Retourner le produit avec la distance minimale
    best_product = df_filtered.loc[df_filtered['distance_to_target'].idxmin()]
    return best_product
```
